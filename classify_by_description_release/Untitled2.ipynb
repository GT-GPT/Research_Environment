```python
import matplotlib.pyplot as plt
```

```python
# data
simple_by_feature_x = ['1', '2', '3', '4', '5', '6', '7']
simple_by_feature_y  = [59.07599926, 59.73200202, 61.63600087, 62.12000251, 62.15199828,
                        62.34200001, 62.34999895]
complex_by_feature_x = ['2', '3', '4', '5']
complex_by_feature_y = [60.42000055, 61.65400147, 62.06799746, 62.00600266]

role_rule_rolerule_x = ['role', 'rule', 'both']
role_rule_rolerule_y = [62.69199848, 62.58000135, 62.53600121]
simple_by_object_x = ['n=1', 'n=2', 'n=3']
simple_by_object_y = [62.58, 62.34999895, 62.62]
```

```python
plt.xlabel('|D(c)|')
plt.ylabel('Top-1 accuracy(%)')

plt.plot(simple_by_feature_x, simple_by_feature_y, marker='o', markersize=5, label = 'simple',  c = 'C2')

plt.plot(complex_by_feature_x, complex_by_feature_y, marker='o', markersize=5, label = 'detailed', c = 'C0')

plt.scatter([4,4.3,4.6],role_rule_rolerule_y,marker=(5, 1),c='orange',label='additional instruction')
plt.annotate("role", (3.6, role_rule_rolerule_y[0]))
plt.annotate("rule", (4.1, role_rule_rolerule_y[1]+0.1))
plt.annotate("role+rule", (4.7, role_rule_rolerule_y[2]))

plt.plot(['1','2','3'],role_rule_rolerule_y,marker=(5, 1),markersize=5,c='purple',label='# of objects in prompt')
plt.annotate("n=1", (-0.1, simple_by_object_y[0]-0.1))
plt.annotate("n=2", (0.8, simple_by_object_y[1]))
plt.annotate("n=3", (1.8, simple_by_object_y[2]-0.3))

plt.legend(loc ='lower right')

plt.savefig('acc.png')
```
